```python
import pandas as pd

df = pd.read_csv('titanic14.csv')
df.head()
```

```python
# simply the dataframe by dropping irrelevant columns
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns', inplace=True)
df.head()
```

```python
# extract target y and features X
y = df.Survived
X = df.drop('Survived', axis='columns')
```

```python
# convert the text ('male', 'female')  in column 'Sex' to numeric
dummies = pd.get_dummies(X.Sex)
X = pd.concat([X, dummies], axis='columns')
X.drop('Sex', axis='columns', inplace=True)
X.head()
```

```python
# check if any column has NA data
X.columns[X.isna().any()]
```

```python
# show some NA data
X.Age[:10]
```

```python
# replace NA with the mean of the column
age_mean = X.Age.mean()
X.Age = X.Age.fillna(age_mean)
X.head(10)
```

```python
# split data into training and test tests
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# create a Naive Bayes model
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
